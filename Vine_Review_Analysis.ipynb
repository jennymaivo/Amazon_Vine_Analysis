```python
import pandas as pd
```

```python
# read vine_table
vine_table_df = pd.read_csv("./vine_table.csv")
vine_table_df
```

```python
vine_table_df.dtypes
```

```python
"""
Filter the data and create a new DataFrame or table to retrieve all the rows where the total_votes count is equal to or greater than 20 to pick reviews that are more likely to be helpful and to avoid having division by zero errors later on.
"""

vine_table_df_20 = vine_table_df.loc[vine_table_df['total_votes']>=20]
vine_table_df_20
```

```python
"""
Filter the new DataFrame or table created in Step 1 and create a new DataFrame or table to retrieve all the rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%.
"""
vine_table_df_20_filter = vine_table_df_20.loc[(vine_table_df_20['helpful_votes'] / vine_table_df_20['total_votes']) >= .50]
vine_table_df_20_filter
```

```python
"""
Filter the DataFrame or table created in Step 2, and create a new DataFrame or table that retrieves all the rows where a review was written as part of the Vine program (paid), vine == 'Y'.
"""
vine_table_df_20_filter_Y = vine_table_df_20_filter.loc[vine_table_df_20_filter['vine'] == "Y"]
vine_table_df_20_filter_Y
```

```python
"""
Repeat Step 3, but this time retrieve all the rows where the review was not part of the Vine program (unpaid), vine == 'N'.
"""

vine_table_df_20_filter_N = vine_table_df_20_filter.loc[vine_table_df_20_filter['vine'] == "N"]
vine_table_df_20_filter_N
```

Determine the total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews for the two types of review (paid vs unpaid).

```python
# Total number of views for paid 
print(f'Total number of views for paid: {len(vine_table_df_20_filter_Y)}')

# Total number of views for unpaid
print(f'Total number of views for unpaid: {len(vine_table_df_20_filter_N)}')
```

```python
# number of 5-star reviews for paid
paid = len(vine_table_df_20_filter_Y[vine_table_df_20_filter_Y['star_rating'] == 5])
print(f'Number of 5-star reviews for paid: {paid}')
        
# number of 5-star reviews for unpaid
unpaid = len(vine_table_df_20_filter_N[vine_table_df_20_filter_N['star_rating'] == 5])
print(f'Number of 5-star reviews for unpaid: {unpaid}')
```

```python
# percentage of 5-star reviews for paid
#(len(vine_table_df_20_filter_Y[vine_table_df_20_filter_Y['star_rating'] == 5]) / len(vine_table_df_20_filter_Y)) * 100 
# no output because of division by 0. There are no paid vine reviews. 

# percentage of 5-star reviews for unpaid 
perc_unpaid = (len(vine_table_df_20_filter_N[vine_table_df_20_filter_N['star_rating'] == 5]) / len(vine_table_df_20_filter_N)) * 100 
print(f'percentage of 5-star reviews for unpaid: {perc_unpaid}')
```
